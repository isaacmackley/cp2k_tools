```python
from cp2k_utils.pdos import pdos_plot
```

The PDOS plotting function **requires** the following inputs:
- A list of the elements in the system **in order of their 'kind'** e.g. Hf is k1, O is k2, ...
  
The following are also configurable, but have defaults if not used:
- Sigma value, used for adjusting the smearing of the plot (default 0.003)
- Value to specify vertical line (default is at 0)
- Whether to include a grid (default is on)
- Minimum x value (default -3)
- Maximum x value (default 6)
- Figure width (default 25)
- Figure height (default 10)

The function will output a plot for each element, showing contributions from different sub orbitals (s, p, d, ...) and also a plot showing all elements with the total contribution from each element.

```python
pdos_plot(['Hf','O','N','H'], sigma=0.0015, vbm=0, grid=False, xmin=-1, xmax=7, figw=15, figh=10)
```
