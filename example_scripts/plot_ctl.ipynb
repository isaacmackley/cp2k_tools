The format of the input data is shown in the cell below.
For each defect, the following is needed.
- Defect name e.g 'N', 'O', 'O2'
- Charge state
- Energy of system with defect
- Chemical potential of added elements
- Charge correction (eV)
- Defect type: 'i' - interstitial, 'v' - vacancy, 's' - substitution
- Site name (leave blank if not required) - useful for differentiating between different sites in a system e.g. a 3-coordinated O site and a 4-coordinated O site
- Chemical potential of removed elements

The function also requires the total energy, VBM and CBM of the pristine system.


**All energies should be given in Hartrees EXCEPT the charge correction, which should be in eV.**
